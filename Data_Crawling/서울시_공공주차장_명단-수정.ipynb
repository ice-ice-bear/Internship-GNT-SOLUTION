```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataset/전국주차장정보표준데이터.csv", encoding="cp949")
df.shape
```

```python
df.head(10)
```

```python
public = df[df['주차장구분'] == '공영']
```

```python
public.info()
```

```python
public.head()
```

```python
# 구군 파생변수 생성
public['소재지도로명주소'].unique
```

```python
public['소재지지번주소'].unique
```

```python
public["시도"] = public["소재지지번주소"].str.split(expand=True)[0]
```

```python
public['시도'] = public['시도'].fillna(public["소재지도로명주소"].str.split(expand=True)[0])
```

```python
public["시도"].unique()
```

```python
#이상치 확인
# 모두 서울 소재지 아님
public.to_csv("results/public.csv", index = False)
```

```python
public_seoul = public[public["시도"] == "서울특별시"]
```

```python
public_seoul.info()
```

```python
public_seoul["구군"] = public_seoul["소재지지번주소"].str.split(expand=True)[1]
```

```python
public_seoul['구군'] = public['구군'].fillna(public["소재지도로명주소"].str.split(expand=True)[1])
```
