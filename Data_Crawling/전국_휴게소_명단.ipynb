```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import lxml

URL = "http://data.ex.co.kr/openapi/restinfo/hiwaySvarInfoList?key=6958898588&type=xml"

# API로 데이터 불러오기
rq = requests.get(URL)
soup = BeautifulSoup(rq.text, "xml")
    
gate_List = []

for item in soup.find_all("list"):
    svarNm = item.find("svarNm").text
    svarGsstClssNm = item.find("svarGsstClssNm").text
    svarGsstClssCd = item.find("svarGsstClssCd").text
    svarCd = item.find("svarCd").text
    svarAddr = item.find("svarAddr").text
    rprsTelNo = item.find("rprsTelNo").text
    routeNm = item.find("routeNm").text
    routeCd = item.find("routeCd").text
    pstnoCd = item.find("pstnoCd").text
    mtnofNm = item.find("mtnofNm").text
    mtnofCd = item.find("mtnofCd").text
    hdqrNm = item.find("hdqrNm").text
    hdqrCd = item.find("hdqrCd").text
    gate_List.append([svarNm, svarGsstClssNm, svarGsstClssCd, svarCd, rprsTelNo, routeNm,
                                                routeCd, pstnoCd, mtnofNm, mtnofCd, hdqrNm, hdqrCd])
        
rest = pd.DataFrame(gate_List, columns= ["svarNm", "svarGsstClssNm", "svarGsstClssCd", "svarCd", "rprsTelNo", "routeNm",
                                                "routeCd", "pstnoCd", "mtnofNm", "mtnofCd", "hdqrNm", "hdqrCd"])

rest.head()
```
