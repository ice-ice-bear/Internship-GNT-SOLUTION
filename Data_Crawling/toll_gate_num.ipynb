```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import lxml


URL = "http://data.ex.co.kr/openapi/basicinfo/updownIcList?key=6958898588&type=xml&numOfRows=99&pageNo=5"
# for i in range(5):
#     URL = "https://data.ex.co.kr/openapi/basicinfo/updownIcList?key=6958898588&type=xml&numOfRows=99&pageNo="+str(i+1)
    
    # API로 데이터 불러오기
rq = requests.get(URL)
soup = BeautifulSoup(rq.text, "xml")
    
gate_List = []

for item in soup.find_all("upDownIcLists"):
    routeName = item.find("routeName").text
    routeNo = item.find("routeNo").text
    unitCode = item.find("unitCode").text
    unitName = item.find("unitName").text
    gate_List.append([routeName, routeNo, unitCode, unitName])
        
gate_df_05 = pd.DataFrame(gate_List, columns= ["routeName", "routeNo", "unitCode", "unitName"])

gate_df_05.head()
```
