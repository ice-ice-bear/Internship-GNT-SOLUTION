```python
import pandas as pd
import os

def extract_excel_info(date):
    
    default_path = f'R:/seatbelt/highway_send_data/{date}_base64'

    txt_path = f'R:/seatbelt/highway_send_data/{date}_base64/plate_0to6/ {date}_plate_0to6.txt'

    excel_path = r'C:/Users/GnTRnD/Desktop/seatbelt.xlsx'

    # Read the text file with the list of filenames
    with open( txt_path , 'r') as file:
        file_names = [line.strip() for line in file]

    # Rename the files
    for filenames in file_names:
        filenames = filenames.split("_")[0:3]
        filenames = "_".join(filenames)


    # Read the Excel file
    excel_data = pd.read_excel( excel_path , sheet_name=f'dp_{date}')

    # Filter rows based on matching filenames
    matching_rows = excel_data[excel_data['filename'].isin(filenames)]

    # Extract the desired columns
    extracted_data = matching_rows[['filename', 'seated']]

    file_name = f'0to6_{date}.csv'

    save_path = os.path.join(default_path, file_name)
    
    # Save the extracted data to a CSV file
    extracted_data.to_csv(save_path , index=False)

    print("Extraction completed. The data has been saved to 'extracted_data.csv'.")

extract_excel_info(230502)
```
