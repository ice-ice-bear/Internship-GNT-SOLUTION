## 1. Sort in ascending number 오름차순 정렬

```python
import os

folder_path = "C:/Users/GnTRnD/Desktop/temp_file/k-city_testset/set_17/labels"

for file_name in os.listdir(folder_path):
    if file_name.endswith(".txt"):
        file_path = os.path.join(folder_path, file_name)
        
        # Read the lines from the file
        with open(file_path, "r") as f:
            lines = f.readlines()
        
        try:
            # Sort the lines based on the first number in each line
            lines = sorted(lines, key=lambda x: int(x.split()[0]))
        except ValueError:
            # If there's aw ValueError, print the file name and continue with the next file
            print(f"ValueError in file {file_name}")
            continue
        
        # Write the sorted lines back to the file
        with open(file_path, "w") as f:
            for line in lines:
                f.write(line.strip() + "\n")
```

## make dataframe of all the detection label

```python
import os
import pandas as pd

folder_path = "C:/Users/GnTRnD/Desktop/temp_file/k-city_testset/set_17/labels"

output_file = folder_path + '/0.class_file.csv'

# Create an empty dataframe to store the results
columns = ['filename'] + list(range(12))
df = pd.DataFrame(columns=columns)

for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        try:
            # Extract the file information
            file_data = []
            with open(os.path.join(folder_path, filename), 'r') as file:
                for line in file:
                    values = line.split()
                    class_number = int(values[0])
                    confidence = float(values[5])
                    file_data.append((class_number, confidence))

            # Create a row for the dataframe
            row = [filename] + ['']*12
            for class_number, confidence in file_data:
                row[class_number+1] = class_number
#                 row[class_number] = confidence
            df = df.append(pd.DataFrame([row], columns=columns), ignore_index=True)
            
        except PermissionError:
            print(f"Permission denied: {filename}")

# Save the dataframe to a CSV file
df.to_csv(output_file, index=False)
```

```python
import os
import pandas as pd

folder_path = "C:/Users/GnTRnD/Desktop/temp_file/k-city_testset/set_17/labels"

output_file = folder_path + '/0.class_file.csv'

# Update the column names to include "_conf" suffix
# Update the columns list
columns = []
for i in range(12):
    columns.append(str(i))
    columns.append(str(i) + '_conf')
    
clumns = ['']

columns = ['filename'] + columns


df = pd.DataFrame(columns=columns)

for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        try:
            # Extract the file information
            file_data = []
            with open(os.path.join(folder_path, filename), 'r') as file:
                for line in file:
                    values = line.split()
                    class_number = int(values[0])
                    confidence = float(values[5])
                    file_data.append((class_number, confidence))

            # Create a row for the dataframe
            row = [filename] + [''] * (len(columns) - 1)
            for class_number, confidence in file_data:
                # Add class number and confidence to the row
                class_col_idx = columns.index(str(class_number))
                conf_col_idx = columns.index(str(class_number) + '_conf')
                row[class_col_idx] = class_number
                row[conf_col_idx] = confidence
            df = df.append(pd.DataFrame([row], columns=columns), ignore_index=True)
            
        except PermissionError:
            print(f"Permission denied: {filename}")

replace_map = {2: 'X', 3: 'O', 4: 'X', 5: 'O', 6: 'X',7: 'O', 8: 'emp', 9: 'emp', 11:'p6'}
df.replace(replace_map, inplace=True)

# Save the dataframe to a CSV file
df.to_csv(output_file, index=False)
```

### Ver1 extracting higest confidence score

```python
import os
import pandas as pd

folder_path = "C:/Users/GnTRnD/Desktop/temp_file/Automation_testset"
output_file = folder_path + '/0.group_file.csv'

# Create an empty dataframe to store the results
columns = ['filename', 'group1', 'group2', 'group3']
df = pd.DataFrame(columns=columns)

for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        try:
            # Extract the file information
            file_data = {}
            with open(os.path.join(folder_path, filename), 'r') as file:
                for line in file:
                    values = line.split()
                    class_number = int(values[0])
                    confidence = float(values[5])
                    file_data[class_number] = confidence

            # Get the highest confidence values for each group
            group1_values = [file_data.get(c, 0) for c in [2, 3, 8]]
            group2_values = [file_data.get(c, 0) for c in [4, 5, 9]]
            group3_values = [file_data.get(c, 0) for c in [6, 7, 10]]
            max_group1 = max(group1_values)
            max_group2 = max(group2_values)
            max_group3 = max(group3_values)

            # Create a row for the dataframe
            row = [filename, max_group1, max_group2, max_group3]
            df = df.append(pd.DataFrame([row], columns=columns), ignore_index=True)
            
        except PermissionError:
            print(f"Permission denied: {filename}")

# Save the dataframe to a CSV file
df.to_csv(output_file, index=False)
```

```python
import os
import pandas as pd

folder_path = "C:/Users/GnTRnD/Desktop/temp_file/Automation_testset"
output_file = folder_path + '/0.group_class_file.csv'

# Create an empty dataframe to store the results
columns = ['filename', 'group1', 'group2', 'group3']
df = pd.DataFrame(columns=columns)

for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        try:
            # Extract the file information
            file_data = {}
            with open(os.path.join(folder_path, filename), 'r') as file:
                for line in file:
                    values = line.split()
                    class_number = int(values[0])
                    confidence = float(values[5])
                    file_data[class_number] = (confidence, values[1])

            # Get the class with the highest confidence value for each group
            group1_values = [(file_data.get(c, (0, None))[0], c) for c in [2, 3, 8]]
            group2_values = [(file_data.get(c, (0, None))[0], c) for c in [4, 5, 9]]
            group3_values = [(file_data.get(c, (0, None))[0], c) for c in [6, 7, 10]]
            max_group1 = max(group1_values)[1] if max(group1_values)[0] != 0 else ''
            max_group2 = max(group2_values)[1] if max(group2_values)[0] != 0 else ''
            max_group3 = max(group3_values)[1] if max(group3_values)[0] != 0 else ''

            # Create a row for the dataframe
            row = [filename, max_group1, max_group2, max_group3]
            df = df.append(pd.DataFrame([row], columns=columns), ignore_index=True)

        except PermissionError:
            print(f"Permission denied: {filename}")

# Save the dataframe to a CSV file
df.to_csv(output_file, index=False)
```

## Make assesment by the given condition

```python
import pandas as pd

folder_path = 'C:/Users/GnTRnD/Desktop/temp_file/Automation_testset'

# Open the CSV file and create a dataframe
df = pd.read_csv(folder_path + '/0.group_answer_sheet.csv')

# Create empty columns for the answers
df['answer1'] = ''
df['answer2'] = ''
df['answer3'] = ''

# Loop through the dataframe in groups of three rows
for i in range(0, len(df), 3):
    # Get the values from each row in the group
    row1 = df.iloc[i]
    row2 = df.iloc[i+1]
    row3 = df.iloc[i+2]
            
    # Determine the most common value in each group and update the answer columns
    for j, group in enumerate(['group1', 'group2', 'group3']):
        values = [row1[group], row2[group], row3[group]]
        o_count = values.count('O')
        x_count = values.count('X')
        if o_count == 0 and x_count == 0:
            df.at[i, 'answer'+str(j+1)] = ''
        elif x_count >= o_count:
            df.at[i, 'answer'+str(j+1)] = 'X'
        else:
            df.at[i, 'answer'+str(j+1)] = 'O'


# Save the updated dataframe to a new CSV file
df.to_csv(folder_path + '/0.answer_sheet_total.csv', index=False)
```

```python
import pandas as pd

folder_path = 'path/to/folder'

# Open the CSV file and create a dataframe
df = pd.read_csv(folder_path + '/0.group_answer_sheet.csv')

# Create empty columns for the answers
df['answer1'] = ''
df['answer2'] = ''
df['answer3'] = ''

# Loop through the dataframe in groups of three rows
for i in range(0, len(df), 3):
    # Get the values from each row in the group
    row1 = df.iloc[i]
    row2 = df.iloc[i+1]
    row3 = df.iloc[i+2]

    # Determine the most common value in each group and update the answer columns
    for j, group in enumerate(['group1', 'group2', 'group3']):
        values = [row1[group], row2[group], row3[group]]
        if 'emt' in values:
            df.at[i, 'answer'+str(j+1)] = 'emt'
        else:
            o_count = values.count('O')
            x_count = values.count('X')
            if o_count == 0 and x_count == 0:
                df.at[i, 'answer'+str(j+1)] = ''
            elif x_count >= o_count:
                df.at[i, 'answer'+str(j+1)] = 'X'
            else:
                df.at[i, 'answer'+str(j+1)] = 'O'

# Loop through the dataframe in groups of three rows
for i in range(0, len(df), 6):
    # Get the values from each row in the group
    row1 = df.iloc[i]
    row2 = df.iloc[i+1]
    row3 = df.iloc[i+2]
    row4 = df.iloc[i+3]
    row5 = df.iloc[i+4]
    row6 = df.iloc[i+5]

    # Determine the most common value in each group and update the answer columns
    for j, group in enumerate(['group3']):
        values = [row1[group], row2[group], row3[group], row4[group], row5[group], row6[group]]
        if 'emt' in values:
            df.at[i, 'answer3'] = 'emt'
            df.at[i+3, 'answer3'] = 'emt'
            
# Save the updated dataframe to a new CSV file
df.to_csv(folder_path + '/0.answer_sheet_total5.csv', index=False)
```
