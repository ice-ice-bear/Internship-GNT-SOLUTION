```python
import os
import shutil

file_path = 'H:/02.가공데이터/02._AIMMO/to_aimmo/2303/230302_1/230302_1.txt'
copy_path = 'H:/02.가공데이터/02._AIMMO/to_aimmo/2303/230302_1'

# Open the file containing the paths
with open( file_path , 'r', encoding='utf-8') as file:
    paths = file.readlines()

# Remove any whitespace characters from each path
paths = [path.strip() for path in paths]

# Loop through each path
for path in paths:
    # Check if the file exists
    if os.path.isfile(path):
        # Copy the file to the desired folder
        shutil.copy(path, copy_path )
```

```python
# Including the folder

import os
from PIL import Image

# Path to the text file containing folder paths
file_path = 'H:/02.가공데이터/02._AIMMO/to_aimmo/2303/230302_1/230302_1.txt'

# Function to extract images from a folder
def extract_images_from_folder(folder_path, save_path):
    image_list = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # Check if the file is an image file (you can modify this condition if needed)
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
                image_path = os.path.join(root, file)
                image_list.append(image_path)
                # Open the image
                image = Image.open(image_path)

                # Save the image to the specified save path
                save_folder = os.path.join(save_path, os.path.basename(folder_path))
                os.makedirs(save_folder, exist_ok=True)
                save_file_path = os.path.join(save_folder, file)
                image.save(save_file_path)
    return image_list

# Read the file path text file
with open(file_path, 'r') as file:
    folder_paths = file.readlines()

# Remove any trailing newline characters
folder_paths = [path.strip() for path in folder_paths]

# Specify the path to save the extracted images
save_path = 'H:/02.가공데이터/02._AIMMO/to_aimmo/2303/230302_1'

# Extract images from each folder path and save them
all_images = []
for folder_path in folder_paths:
    images = extract_images_from_folder(folder_path, save_path)
    all_images.extend(images)

# Example usage: Print the paths of the extracted images
for image_path in all_images:
    print(image_path)

```

```python
# Not including the folder

import os
import shutil

# Path to the text file containing folder paths
file_path = 'H:/02.가공데이터/02._AIMMO/to_aimmo/2303/230302_1/230302_1.txt'

# Function to move images from a folder to the specified save path
def move_images_from_folder(folder_path, save_path):
    image_list = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # Check if the file is an image file (you can modify this condition if needed)
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')) and "_yolo" not in file.lower():
                image_path = os.path.join(root, file)
                image_list.append(image_path)
                # Move the image to the specified save path
                save_file_path = os.path.join(save_path, file)
                shutil.copy(image_path, save_file_path)
    return image_list

# Read the file path text file
with open(file_path, 'r') as file:
    folder_paths = file.readlines()

# Remove any trailing newline characters
folder_paths = [path.strip() for path in folder_paths]

# Specify the path to move the extracted images
save_path = 'H:/02.가공데이터/02._AIMMO/to_aimmo/2303/230302_1'

# Move images from each folder path to the save path
all_images = []
for folder_path in folder_paths:
    images = move_images_from_folder(folder_path, save_path)
    all_images.extend(images)

# Example usage: Print the paths of the moved images
for image_path in all_images:
    print(image_path)

```
