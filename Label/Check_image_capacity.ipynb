```python
import os

folder_path = "//schoolnas-work/schoolnas-work/seatbelt/dp_daily_testdata/detect/dp_230415/labels"
max_file_size = 0  # 300 KB in bytes

# Step 1: Find images that are less than max_file_size
keywords = set()
for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)

    if os.path.isfile(file_path):
        file_size = os.path.getsize(file_path)
        if file_size <= max_file_size:
            try:
                keyword = file_name.split("_")[2]
                keywords.add(keyword)
            except IndexError:
                print(f"Error: Could not extract keyword from {file_name}")

# Step 2: Remove images with the same keyword
for keyword in keywords:
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)

        if os.path.isfile(file_path):
            if keyword in file_name:
                os.remove(file_path)
```
