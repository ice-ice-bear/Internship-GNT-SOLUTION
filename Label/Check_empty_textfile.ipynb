## with the file extension

```python
import os

# set the path of the folder containing text files
folder_path = "//schoolnas-work/schoolnas-work/seatbelt/aimmo_230323_labels/gnt_solution_output/gnt_solution_output/대왕판교선별이미지(221229_230110)/230104"

# create an empty list to store the names of empty text files
empty_files = []

# loop through each file in the folder
for file_name in os.listdir(folder_path):
    # check if the file is a text file
    if file_name.endswith(".txt"):
        # get the full path of the file
        file_path = os.path.join(folder_path, file_name)
        # check if the file is empty
        if os.path.getsize(file_path) == 0:
            # add the file name to the empty_files list
            empty_files.append(file_name)
            # print the file name
            print(file_name)
            
# write the list of empty text files to a file
with open(os.path.join(folder_path, "0.empty_files.txt"), "w") as f:
    for file_name in empty_files:
        f.write(file_name + "\n")
```

## without the file extension

```python
import os

# set the path of the folder containing text files
folder_path = "//schoolnas-work/schoolnas-work/seatbelt/aimmo_230323_labels/gnt_solution_output/gnt_solution_output/대왕판교선별이미지(221229_230110)/230103"

# create an empty list to store the names of empty text files
empty_files = []

# loop through each file in the folder
for file_name in os.listdir(folder_path):
    # check if the file is a text file
    if file_name.endswith(".txt"):
        # get the full path of the file
        file_path = os.path.join(folder_path, file_name)
        # check if the file is empty
        if os.path.getsize(file_path) == 0:
            # add the file name without extension to the empty_files list
            empty_files.append(os.path.splitext(file_name)[0])
            # print the file name
            print(file_name)

# write the list of empty text files to a file
with open(os.path.join(folder_path, "0.empty_files.txt"), "w") as f:
    for file_name in empty_files:
        f.write(file_name + "\n")
```

```python
#Only care about the file name not the file extension
import os
import shutil

image_folder = folder_path
text_file =  image_folder + r'/0.empty_files.txt'
output_folder = image_folder + r'/empty_files'

with open(text_file, 'r') as f:
    exclude_list = set(f.read().splitlines())

for root, dirs, files in os.walk(image_folder):
    for file in files:
        filename = os.path.splitext(file)[0] # get filename without extension
        if filename in exclude_list:
            image_path = os.path.join(root, file)
            output_path = os.path.join(output_folder, file)
            print('Moving file:', image_path, 'to', output_path)
            shutil.move(image_path, output_path)
```
