## sort label within ascending order

```python
import os

folder_path = "//schoolnas-work/schoolnas-work/seatbelt/spark_videos/ay_spark_videos"

for file_name in os.listdir(folder_path):
    if file_name.endswith(".txt"):
        file_path = os.path.join(folder_path, file_name)
        
        # Read the lines from the file
        with open(file_path, "r") as f:
            lines = f.readlines()
        
        try:
            # Sort the lines based on the first number in each line
            lines = sorted(lines, key=lambda x: int(x.split()[0]))
        except ValueError:
            # If there's aw ValueError, print the file name and continue with the next file
            print(f"ValueError in file {file_name}")
            continue
        
        # Write the sorted lines back to the file
        with open(file_path, "w") as f:
            for line in lines:
                f.write(line.strip() + "\n")
```

```python
import os

folder_path = "C:/Users/GnTRnD/Desktop/temp_file/gnt_solution_output (2)/230103 - "

for file_name in os.listdir(folder_path):
    if file_name.endswith(".txt"):
        file_path = os.path.join(folder_path, file_name)
        
        # Read the lines from the file
        with open(file_path, "r") as f:
            lines = f.readlines()
        
        try:
            # Sort the lines based on the first and second numbers in each line
            lines = sorted(lines, key=lambda x: (int(x.split()[0] == '0'), int(x.split()[0]), int(x.split()[1])))
        except ValueError:
            # If there's a ValueError, print the file name and continue with the next file
            print(f"ValueError in file {file_name}")
            continue
        
        # Write the sorted lines back to the file
        with open(file_path, "w") as f:
            f.writelines(lines)
```

## if certain class number pops up more than one time print the file name

```python
#if certain class number pops up more than one time print the file name 
import os

# Define the folder path
folder_path = '//schoolnas-work/schoolnas-work/seatbelt/seatbelt_label_03_06_added_230324_train_data/12_sh_220215(srlee_done)'

# Loop over each file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.txt'):
        # Get the full file path
        file_path = os.path.join(folder_path, file_name)

        # Read the file line by line
        with open(file_path, 'r') as file:
            lines = file.readlines()

        # Count the number of lines starting with '9'
        count = 0
        for line in lines:
            if line.startswith('9'):
                count += 1

        # If there is more than one line starting with '9', print the file name
        if count > 1:
            print(file_name)
```
