```python
import pandas as pd
import os

date = '230727'

folder_path = r'R:\seatbelt\dp_daily_testdata_statistic'

# Replace 'seats.csv' with the actual file path if it's not in the current directory
file_name = f'{date}_230412_seatbelt_traingset06.pt_assesment_datasheet.csv'

file_path = os.path.join(folder_path, file_name)

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, encoding='cp949')

# Function to count occurrences of 'O', 'X', and total for a given column
def count_values(column_name):
    count_O = df[column_name].eq('O').sum()
    count_X = df[column_name].eq('X').sum()
    count_total = count_O + count_X
    return count_O, count_X, count_total

# Column names for each category
seat1_columns = ['seat1']
seat3_columns = ['seat3']
seat_back_columns = ['seat4', 'seat5', 'seat6']

# Initialize dictionaries to store counts for each category
seat1_counts = {}
seat3_counts = {}
seat_back_counts = {}

# Count for 'seat1' category
for column in seat1_columns:
    seat1_counts['seat1_O_count'], seat1_counts['seat1_X_count'], seat1_counts['seat1_total'] = count_values(column)

# Count for 'seat3' category
for column in seat3_columns:
    seat3_counts['seat3_O_count'], seat3_counts['seat3_X_count'], seat3_counts['seat3_total'] = count_values(column)

# Count for 'seat4', 'seat5', and 'seat6' category (summarized)
seat_back_counts['seat_back_O_count'], seat_back_counts['seat_back_X_count'], seat_back_counts['seat_back_total'] = count_values('seat4')[0] + count_values('seat5')[0] + count_values('seat6')[0], count_values('seat4')[1] + count_values('seat5')[1] + count_values('seat6')[1], count_values('seat4')[2] + count_values('seat5')[2] + count_values('seat6')[2]

# Create a DataFrame with the results
data = {
    'seat1_total': [seat1_counts['seat1_total']],
    'seat1_O_count': [seat1_counts['seat1_O_count']],
    'seat1_X_count': [seat1_counts['seat1_X_count']],
    'seat3_total': [seat3_counts['seat3_total']],
    'seat3_O_count': [seat3_counts['seat3_O_count']],
    'seat3_X_count': [seat3_counts['seat3_X_count']],
    'seat_back_total': [seat_back_counts['seat_back_total']],
    'seat_back_O_count': [seat_back_counts['seat_back_O_count']],
    'seat_back_X_count': [seat_back_counts['seat_back_X_count']],
}

df_result = pd.DataFrame(data, index=[date])

# Print the resulting DataFrame
print(df_result)
```
