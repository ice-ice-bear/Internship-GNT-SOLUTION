```python
import os
import pandas as pd

df_lr['detect1'] = ''
df_lr['detect2'] = ''

for i in range(0, len(df_lr), 3):
    row1 = df_lr.iloc[i]
    row2 = df_lr.iloc[i+1]
    row3 = df_lr.iloc[i+2]

    for j, k in enumerate(['01_l']):
        values = [row1[k]]
        df_lr.at[i+2, 'detect1'] = values[0]
    
    for j, k in enumerate(['01_r']):
        values = [row1[k]]
        df_lr.at[i, 'detect1'] = values[0]
        
    for j, k in enumerate(['04_l']):
        values = [row2[k], row3[k]]
        o_count = values.count('O')
        x_count = values.count('X')
        emp_count = values.count('emp')
        if o_count == 0 and x_count == 0:
            df_lr.at[i+2, 'detect2'] = ''
        elif emp_count == 2:
            df_lr.at[i+2, 'detect2'] = ''
        elif o_count >= x_count:
            df_lr.at[i+2, 'detect2'] = 'O'
        else:
            df_lr.at[i+2, 'detect2'] = 'X'
            
    
    for j, k in enumerate(['04_r']):
        values = [row2[k], row3[k]]
        o_count = values.count('O')
        x_count = values.count('X')
        emp_count = values.count('emp')
        if o_count == 0 and x_count == 0:
            df_lr.at[i, 'detect2'] = ''
        elif emp_count == 2:
            df_lr.at[i, 'detect2'] = ''
        elif o_count >= x_count:
            df_lr.at[i, 'detect2'] = 'O'
        else:
            df_lr.at[i, 'detect2'] = 'X'
            
    
    for j, k in enumerate(['05_l', '05_r']):
        values_l = [row2['05_l'], row3['05_l']]
        values_r = [row2['05_r'], row3['05_r']]
        o_count_l = values_l.count('O') 
        x_count_l = values_l.count('X')
        emp_count_l = values_l.count('emp')
        o_count_r = values_r.count('O') 
        x_count_r = values_r.count('X')
        emp_count_r = values_r.count('emp')

        if o_count_l == 0 and x_count_l == 0:
            l_count = ''
        elif emp_count_l == 2:
            l_count = ''
        elif o_count_l >= x_count_l:
            l_count = 'O'
        else:
            l_count = 'X'

        if o_count_r == 0 and x_count_r == 0:
            r_count = ''
        elif emp_count_r == 2:
            r_count = ''
        elif o_count_r >= x_count_r:
            r_count = 'O'
        else:
            r_count = 'X'

        if l_count == r_count:
            df_lr.at[i+1, 'detect2'] = l_count
        elif l_count == '' and r_count == '':
            df_lr.at[i+1, 'detect2'] = ''
        elif l_count == 'O' and r_count == 'X':
            df_lr.at[i+1, 'detect2'] = 'O'
        elif l_count == 'X' and r_count == 'O':
            df_lr.at[i+1, 'detect2'] = 'O'
        else:
            df_lr.at[i+1, 'detect2'] = 'X'
            
    for j, k in enumerate(['06_l', '06_r']):
        values = [row2[k], row3[k]]
        if 'p6' in values:
            df_lr.at[i+1, 'detect2'] = ''
            
replace_map = {'emp' : ''}
df_lr['detect1'].replace(replace_map, inplace=True)
df_lr['detect2'].replace(replace_map, inplace=True)

df_lr.to_csv(folder_path + '/Assesment_datasheet.csv',index=False)
```

```python
import os
import pandas as pd

folder_path = "C:/Users/GnTRnD/Desktop/temp_file/k-city_testset/set_17/labels"

output_file = folder_path + '/0.class_file.csv'

# Update the column names to include "_conf" suffix
# Update the columns list
columns = []
for i in range(12):
    columns.append(str(i))
    columns.append(str(i) + '_conf')
    
clumns = ['']

columns = ['filename'] + columns


df = pd.DataFrame(columns=columns)

for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        try:
            # Extract the file information
            file_data = []
            with open(os.path.join(folder_path, filename), 'r') as file:
                for line in file:
                    values = line.split()
                    class_number = int(values[0])
                    confidence = float(values[5])
                    file_data.append((class_number, confidence))

            # Create a row for the dataframe
            row = [filename] + [''] * (len(columns) - 1)
            for class_number, confidence in file_data:
                # Add class number and confidence to the row
                class_col_idx = columns.index(str(class_number))
                conf_col_idx = columns.index(str(class_number) + '_conf')
                row[class_col_idx] = class_number
                row[conf_col_idx] = confidence
            df = df.append(pd.DataFrame([row], columns=columns), ignore_index=True)
            
        except PermissionError:
            print(f"Permission denied: {filename}")

replace_map = {2: 'X', 3: 'O', 4: 'X', 5: 'O', 6: 'X',7: 'O', 8: 'emp', 9: 'emp', 11:'p6'}
df.replace(replace_map, inplace=True)

# Save the dataframe to a CSV file
df.to_csv(output_file, index=False)
```
