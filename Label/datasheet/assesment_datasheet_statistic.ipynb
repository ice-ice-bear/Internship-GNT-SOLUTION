```python
import pandas as pd

def count_seats(data, hour_start, hour_end):
    filtered_data = data[(data['filename'].dt.hour >= hour_start) & (data['filename'].dt.hour < hour_end)]
    
    o_counts = []
    x_counts = []
    total_counts = [] 
    
    for seat_num in [1, 3, 4, 5, 6]:
        o_count = (filtered_data[f'seat{seat_num}'] == 'O').sum()
        x_count = (filtered_data[f'seat{seat_num}'] == 'X').sum()
        total_count = o_count + x_count
        
        o_counts.append(o_count)
        x_counts.append(x_count)
        total_counts.append(total_count)
    
    return o_counts, x_counts, total_counts

date = '230821'

file_name = f'{date}_dataset06_assesment_datasheet.csv'

data = pd.read_csv(file_name, encoding='cp949')
data['filename'] = pd.to_datetime(data['filename'], format='%y%m%d_%H%M%S')

time_ranges = [(7, 9), (18, 20), (20, 22)]
seat_labels = ['seat1', 'seat3', 'seat4', 'seat5', 'seat6']

results = []

for hour_start, hour_end in time_ranges:
    o_counts, x_counts, total_counts = count_seats(data, hour_start, hour_end)
    results.append(o_counts + x_counts + total_counts)

column_labels = [f'{seat}_O' for seat in seat_labels] + [f'{seat}_X' for seat in seat_labels] + [f'{seat}_total' for seat in seat_labels]
ox_count = pd.DataFrame(results, columns=column_labels)
ox_count.index = ['7to9', '18to20', '20to22']

refined_colunm = []
for i in [1, 3, 4, 5, 6]:
    refined_colunm.extend([f'seat{i}_total', f'seat{i}_O', f'seat{i}_X'])

ox_count = ox_count[refined_colunm]
```
