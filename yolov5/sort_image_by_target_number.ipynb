```python
import re

# Define the regex pattern to match the number between the second and third underscore
pattern = re.compile(r'.*_\d+_(\d+)_.*')

# Define the path of the text file containing the image names
image_names_file = './suv_and_carnival_sorted_results.txt'

# Define the path where the target numbers will be saved
target_numbers_file = './suv_and_carnival_target_number.txt'

# Create a set to store the processed target numbers
processed_target_numbers = set()

# Open the target numbers file in append mode
with open(target_numbers_file, 'a') as f:
    # Open the image names file in read mode
    with open(image_names_file, 'r') as f_image_names:
        # Loop through all the image names in the file
        for filename in f_image_names:
            # Remove the newline character at the end of the filename
            filename = filename.strip()
            # Check if the filename matches the pattern
            match = pattern.match(filename)
            if match:
                # Extract the number from the filename
                number = match.group(1)
                # Check if the number has already been processed
                if number in processed_target_numbers:
                    print(f"Skipped {filename} because it has the same target number as a previous file")
                else:
                    # If the number hasn't been processed, write it to the target numbers file and add it to the set of processed target numbers
                    f.write(f"{number}\n")
                    processed_target_numbers.add(number)
```

```python
import os
import shutil

# Define the path of the directory containing the images
images_dir = '/path/to/images'

# Define the path of the text file containing the target numbers
target_numbers_file = '/path/to/target_numbers.txt'

# Define the path of the directory where the matched images will be moved to
matched_images_dir = '/path/to/matched_images'

# Create the matched images directory if it doesn't exist
if not os.path.exists(matched_images_dir):
    os.makedirs(matched_images_dir)

# Open the target numbers file in read mode
with open(target_numbers_file, 'r') as f:
    # Get a list of all target numbers in the file
    target_numbers = [number.strip() for number in f]

# Loop through all the files in the images directory
for filename in os.listdir(images_dir):
    # Check if the filename contains any target number
    matched = False
    for number in target_numbers:
        if number in filename:
            # If it does, move the file to the matched images directory
            src_path = os.path.join(images_dir, filename)
            dest_path = os.path.join(matched_images_dir, filename)
            shutil.move(src_path, dest_path)
            matched = True
            break
    # If the filename doesn't contain any target number, print a message
    if not matched:
        print(f"Skipping {filename} as it doesn't match any target number")
```
